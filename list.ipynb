```python
# List


# Creating a list
my_list = [1, 2, 3, 4, 5]

# 1. append() - Adds an element to the end of the list
my_list.append(6)
print("After append(6):", my_list)

# 2. extend() - Adds multiple elements to the end of the list
my_list.extend([7, 8, 9])
print("After extend([7, 8, 9]):", my_list)

# 3. insert() - Inserts an element at a specified position
my_list.insert(0, 5)  # Insert 0 at index 0
print("After insert(0, 0):", my_list)

# 4. remove() - Removes the first occurrence of a specified value
my_list.remove(3)
print("After remove(3):", my_list)

# 5. pop() - Removes and returns the element at the specified position
popped_value = my_list.pop(2)  # Pop the element at index 2
print("Popped value:", popped_value)
print("After pop(2):", my_list)

# 6. index() - Returns the index of the first occurrence of a specified value
index_of_5 = my_list.index(5)
print("Index of 5:", index_of_5)

# 7. count() - Returns the number of occurrences of a specified value
count_of_1 = my_list.count(1)
print("Count of 1:", count_of_1)

# 8. sort() - Sorts the list in ascending order
my_list.sort()
print("After sort():", my_list)

# 9. min() - Finds the minimum value in the list
min_value = min(my_list)
print("Minimum value:", min_value)

# 10. max() - Finds the maximum value in the list
max_value = max(my_list)
print("Maximum value:", max_value)

# 11. reverse() - Reverses the order of the list
my_list.reverse()
print("After reverse():", my_list)

# 12. clear() - Removes all elements from the list
my_list.clear()
print("After clear():", my_list)
```

```python
# Demonstrating various dictionary methods in Python

# Creating a dictionary
my_dict = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

# Display the original dictionary
print("Original dictionary:", my_dict)

# 1. get() - Retrieves the value for a specified key
name = my_dict.get('name')
print("Value of 'name':", name)

# 2. keys() - Returns a view object that displays a list of all the keys
keys = my_dict.keys()
print("Keys in the dictionary:", keys)

# 3. values() - Returns a view object that displays a list of all the values
values = my_dict.values()
print("Values in the dictionary:", values)

# 4. items() - Returns a view object that displays a list of the dictionary's key-value pairs
items = my_dict.items()
print("Key-Value pairs in the dictionary:", items)

# 5. update() - Updates the dictionary with elements from another dictionary or iterable
my_dict.update({'age': 31, 'country': 'USA'})
print("After update():", my_dict)

del my_dict['country']
print(my_dict)  

# 6. pop() - Removes the specified key and returns its value
popped_value = my_dict.pop('city')
print("Popped value for 'city':", popped_value)
print("After pop('city'):", my_dict)

# 7. popitem() - Removes and returns the last inserted key-value pair as a tuple
last_item = my_dict.popitem()
print("Popped last item:", last_item)
print("After popitem():", my_dict)



# 8. clear() - Removes all items from the dictionary
my_dict.clear()
print("After clear():", my_dict)

# 9. copy() - Returns a shallow copy of the dictionary
my_dict = {'name': 'Alice', 'age': 30}
copy_dict = my_dict.copy()
print("Original dictionary:", my_dict)
print("Copied dictionary:", copy_dict)
```

```python
# dictionary in list 

# Creating an initial list
my_list = [1, 2, 3]

# 1. append() - Adds an element to the end of the list
my_list.append(4)
print("After append(4):", my_list)

# 2. extend() - Adds multiple elements to the end of the list
my_list.extend([5, 6])
print("After extend([5, 6]):", my_list)

# 3. reverse() - Reverses the order of the list
my_list.reverse()
print("After reverse():", my_list)

# 4. pop() - Removes and returns the element at the specified position (last by default)
popped_value = my_list.pop()  # Popping the last element
print("Popped value:", popped_value)
print("After pop():", my_list)

# 5. remove() - Removes the first occurrence of a specified value
my_list.remove(2)
print("After remove(2):", my_list)

# 6. copy() - Returns a shallow copy of the list
copied_list = my_list.copy()
print("Copied list:", copied_list)

# 7. Adding more elements for demonstration
my_list.append(7)
my_list.append(8)
print("After adding 7 and 8:", my_list)

# 8. Reverse the copied list
copied_list.reverse()
print("Reversed copied list:", copied_list)

# Final output of both lists
print("Final my_list:", my_list)
print("Final copied_list:", copied_list)
```

```python
# Set methods

my_set = {1, 2, 3, 4, 5}
print("Original set:", my_set)

# 1. add() - Adds an element to the set
my_set.add(6)
print("After add(6):", my_set)

# 2. update() - Adds multiple elements to the set
my_set.update([7, 8, 9])
print("After update([7, 8, 9]):", my_set)

# 3. remove() - Removes a specified element; raises KeyError if not found
my_set.remove(4)
print("After remove(4):", my_set)

# 4. discard() - Removes a specified element; does not raise an error if not found
my_set.discard(10)  # No error, since 10 is not in the set
print("After discard(10):", my_set)

# 5. pop() - Removes and returns an arbitrary element from the set
popped_value = my_set.pop()  # Random element removed
print("Popped value:", popped_value)
print("After pop():", my_set)

# 6. clear() - Removes all elements from the set
my_set.clear()
print("After clear():", my_set)

# 7. Creating another set for demonstration
another_set = {3, 4, 5, 6, 7}

# 8. union() - Returns a new set with elements from both sets
union_set = my_set.union(another_set)
print("Union of my_set and another_set:", union_set)

# 9. intersection() - Returns a new set with elements common to both sets
intersection_set = my_set.intersection(another_set)
print("Intersection of my_set and another_set:", intersection_set)

# 10. difference() - Returns a new set with elements in the first set but not in the second
difference_set = another_set.difference(my_set)
print("Difference of another_set - my_set:", difference_set)
```
