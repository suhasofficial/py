```python
#pandas series and their methods 
 
import pandas as pd


# Creating a simple Pandas Series from a list
data = [10, 20, 30, 40]
series = pd.Series(data)

# 1. First 2 elements using head()
print(series.head(2))

# 2. Last 2 elements using tail()
print(series.tail(2))

# 3. Summary statistics using describe()
print(series.describe())

# print("iloc loc")
# print(series.loc[0:4,['name','percetage']])
# print(series.iloc[1:4])

# 4. Data type of the series using dtype
print(series.dtype)

# 5. Index and values of the series using index and values attributes
print(series.index)
print(series.values)

# 6. Sum of the series
print(series.sum())

# 7. Mean of the series
print(series.mean())

# 8. Minimum value in the series using min()
print(series.min())

# 9. Maximum value in the series using max()
print(series.max())

```

```python
# barh graph 

import matplotlib.pyplot as plt

data = {'c':39, 'c++':30, 'java':50, 'python':34}

course = list(data.keys())
values = list(data.values())

plt.barh(course, values, color='purple')
plt.show()
```

```python
# piechart 

import matplotlib.pyplot as plt


labels = ['Apple', 'Samsung', 'Xiaomi', 'Nokia']
values = [10, 20, 30, 40]
colors = ['yellow', 'green', 'orange', 'blue']

plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
#Line chart
 
import matplotlib.pyplot as plt
 
data = {
    'sir1' : [1,3,4,5,6,7 ,8,9],
        'sir2' :[5,7,4,7 ,8,9],
        'sir3' : [6,4,9,5,3,2,5,12],
    }

plt.figure(figsize=(10,6))
plt.plot(data['sir1'], label='sir1', marker='o', color='r')
plt.plot(data['sir2'], label='sir2', marker='^', color='g')
plt.plot(data['sir3'], label='sir3', marker='s', color='b')
plt.xlabel('index')
plt.ylabel('values')
plt.title('desc')
plt.legend()
plt.grid()
plt.show()
```
