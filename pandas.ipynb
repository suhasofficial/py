```python
#pandas series and their methods 
 
import pandas as pd


# Creating a simple Pandas Series from a list
data = [10, 20, 30, 40]
series = pd.Series(data)

# 1. First 2 elements using head()
print(series.head(2))

# 2. Last 2 elements using tail()
print(series.tail(2))

# 3. Summary statistics using describe()
print(series.describe())

# print("iloc loc")
# print(series.loc[0:4,['name','percetage']])
# print(series.iloc[1:4])

# 4. Data type of the series using dtype
print(series.dtype)

# 5. Index and values of the series using index and values attributes
print(series.index)
print(series.values)

# 6. Sum of the series
print(series.sum())

# 7. Mean of the series
print(series.mean())

# 8. Minimum value in the series using min()
print(series.min())

# 9. Maximum value in the series using max()
print(series.max())

```

```python
# aggregate 

import pandas as pd

data = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'marks': [[90, 80, 99, 100], [100, 76, 55, 34], [100, 34, 89, 90], [67, 88, 77, 99]]
}

df = pd.DataFrame(data)
display(df)

# Applying aggregation functions to the 'marks' column
aggregated_marks = df.groupby('ID').agg({
    'marks': lambda x: [pd.Series(mark).agg(['mean', 'min', 'max']) for mark in x]
})

display(aggregated_marks)
```
